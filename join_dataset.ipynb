```python
studentasses = pd.read_csv("dataset/studentAssessment.csv")
```

```python
studentinfo = pd.read_csv("dataset/studentInfo.csv")
```

```python
# from here we start to join all dataset
```

```python
df_student = pd.merge(
    studentinfo,
    studentreg,
    on=["code_module", "code_presentation", "id_student"],
    how="left"
)
```

```python
df_student.head(10)
```

```python
df_student.info()
```

```python
df_student = pd.merge(
    df_student,
    courses,
    on=["code_module", "code_presentation"],
    how="left"
)
```

```python
df_student.head(10)
```

```python
df_student.info()
```

```python
df_assessment = pd.merge(
    studentasses,
    assessment,
    on="id_assessment",
    how="left"
)
```

```python
df_assessment.head(10)
```

```python
df_assessment.info()
```

```python
df_student = pd.merge(
    df_student,
    df_assessment,
    on=["id_student", "code_module", "code_presentation"],
    how="left"
)
```

```python
df_student.head(10)
```

```python
df_student.info()
```

```python
df_vle = pd.merge(
    studentvle,
    vle,
    on=["id_site", "code_module", "code_presentation"],
    how="left"
)
```

```python
df_vle.head(10)
```

```python
df_vle.info()
```

```python
vle_agg = (
    studentvle
    .groupby(["code_module", "code_presentation", "id_student"])
    ["sum_click"]
    .sum()
    .reset_index()
)
```

```python
df_student = pd.merge(
    df_student,
    vle_agg,
    on=["code_module", "code_presentation", "id_student"],
    how="left"
)
```

```python
df_student.head(10)
```

```python
df_student.info()
```

```python
df_student.to_excel("student_dataset.xlsx")
```
