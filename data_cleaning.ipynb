```python
import pandas as pd
```

```python
student = pd.read_excel("student_dataset.xlsx")
```

```python
student.to_csv("student_dataset.csv")
```

```python
student.head(10)
```

```python
# Handle Missing Valuesabs
```

```python
student.isnull().sum()
```

```python
student['id_assessment'] = student['id_assessment'].fillna(
    student['id_assessment'].mode()[0]
)
```

```python
student.isnull().sum()
```

```python
student['date_submitted'] = student['date_submitted'].fillna(student['date_submitted'].median())
```

```python
student.isnull().sum()
```

```python
student['is_banked'] = student['is_banked'].fillna(
    student['is_banked'].mode()[0]
)
student.isnull().sum()
```

```python
student['score'] = pd.to_numeric(student['score'], errors='coerce')
```

```python
# score
student['score'] = student['score'].fillna(student['score'].median())

# assessment_type
student['assessment_type'] = student['assessment_type'].fillna(
    student['assessment_type'].mode()[0]
)

student['date'] = pd.to_numeric(student['date'], errors='coerce')

# date
student['date'] = student['date'].fillna(student['date'].median())

# weight
student['weight'] = student['weight'].fillna(student['weight'].median())

# sum_click
student['sum_click'] = student['sum_click'].fillna(student['sum_click'].median())
student.isnull().sum()
```

```python
# duplicate data 
```

```python
student.duplicated().sum()
```

```python
student.drop_duplicates(inplace=True)
```
